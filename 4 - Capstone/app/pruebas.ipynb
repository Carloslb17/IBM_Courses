```python
import pandas as pd
import numpy as np

# Configuraci√≥n de tiempo (0 a 1 segundo con intervalo de 0.1 segundos)
tiempo = np.arange(0, 1.1, 0.1)

# Generar datos aleatorios para Potencia Babor en RPM, Potencia Estribor en RPM y Kw
np.random.seed(42)  # Para reproducibilidad
potencia_babor = np.random.randint(100, 200, len(tiempo))
potencia_estribor = np.random.randint(50, 150, len(tiempo))
kw = np.random.uniform(50, 150, len(tiempo))

# Crear DataFrame
data = {
    'Tiempo': tiempo,
    'Potencia_Babor_RPM': potencia_babor,
    'Potencia_Estribor_RPM': potencia_estribor,
    'KW': kw
}

df = pd.DataFrame(data)

# Mostrar el DataFrame
print(df)
```

```python
df['Q_babor'] = (df['KW'] * 1000 * 60 / 2 * 3) * (1 / df['Potencia_Babor_RPM'])
df['Q_estribor'] = (df['KW'] * 1000 * 60 / 2 * 3) * (1 / df['Potencia_Estribot_RPM'])
```
